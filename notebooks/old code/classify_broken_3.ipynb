```python
from PIL import Image
from IPython.display import display_png
import pickle
from IPython.display import clear_output


with open('broken_banners.pkl', 'rb') as f:
    broken_heatmaps = pickle.load(f)



for ind, directory in enumerate(heatmap_directories):
    if ind <= 260:
        continue
    _, h, w, g_id, _, domain  = heatmap_file_names[ind][:-4].split("_")
    print("Index: ", ind)
    print(h, w, g_id, domain)

    image_path = heatmap_directories[ind]
    img = Image.open(image_path)
    display_png(img)
    is_broken = input("Is_broken")
    if is_broken == 'y':
        broken_heatmaps.append((h, w, g_id, domain))
        with open('dataset3_broken_banners.pkl', 'wb') as f:
            pickle.dump(broken_heatmaps, f)
    clear_output(wait=True)
```
